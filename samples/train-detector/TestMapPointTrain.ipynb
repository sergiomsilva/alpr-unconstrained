```python
folder =['00029','00011','00024']
path='./'
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '../../train/annotate/' 
in_dir = os.listdir(path)
folder = []
```

```python
for i in in_dir:
    n = i.split('.')[0]
    if n not in folder:
        folder.append(n)
# folder.remove('')
print(folder)
```

```python
for i in folder:
    img = cv2.imread('{}{}.jpg'.format(path,i))
    h,w = img.shape[:2]
    print(w,h)
    data = open('{}{}.txt'.format(path,i),'r').read().split(',')
    for d in range(len(data)):
        if(d==0):
            continue
        elif(d<=int(data[0])):
            data[d] = float(data[d])*w
        elif(d<=int(data[0])*2):
            data[d] = float(data[d])*h
    plt.imshow(img)
    plt.show()
    pts = np.array( [[data[1],data[5]],
                  [data[2],data[6]],
                  [data[3],data[7]],
                  [data[4],data[8]]], np.int32)
    pts = pts.reshape((-1,1,2))
    img2 = cv2.polylines(img,[pts],True,(0,255,0),3)
    cv2.imwrite('out/{}.png'.format(i),img2)
    plt.imshow(img2)
    plt.show()
```
